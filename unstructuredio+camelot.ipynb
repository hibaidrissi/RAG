## Extraction des tableaux du PDF

```python
from unstructured_inference.inference.layout import DocumentLayout

def extract_tables_and_dimensions(fname):
    layout = DocumentLayout.from_file(fname)

    results = {}

    for page in layout.pages:
        tables = [element for element in page.elements if element.type == 'Table']

        if tables:
            page_info = {
                'page_size': f"{page.image_metadata['width']} x {page.image_metadata['height']}",
                'format': page.image_metadata['format'],
                'tables': [{
                    'bbox': {
                        'x1': element.bbox.x1,
                        'y1': element.bbox.y1,
                        'x2': element.bbox.x2,
                        'y2': element.bbox.y2
                    }
                } for element in tables]
            }

            results[page.number] = page_info

    return results

# Usage example
fname = "AircraftPanel Case.pdf"
table_info = extract_tables_and_dimensions(fname)
print(table_info)
```

```python
from pdfminer.pdfpage import PDFPage
from pdfminer.pdfparser import PDFParser
from pdfminer.pdfdocument import PDFDocument

def get_pdf_dimensions(file_path):
    with open(file_path, 'rb') as file:
        parser = PDFParser(file)
        document = PDFDocument(parser)

        page_dimensions = set()  # Use a set to avoid duplicates

        for page in PDFPage.create_pages(document):
            width = page.mediabox[2] - page.mediabox[0]
            height = page.mediabox[3] - page.mediabox[1]
            page_dimensions.add((width, height))  # Add dimensions as a tuple

        # If all pages have the same dimensions, return just one
        if len(page_dimensions) == 1:
            return page_dimensions.pop()  # Return the single set of dimensions
        else:
            return list(page_dimensions)  # Return a list of all unique dimensions

# Usage example
pdf_file_path = "AircraftPanel Case.pdf"
dimensions = get_pdf_dimensions(pdf_file_path)
print(dimensions)
```

```python
def convert_bbox_to_pdf_coords(tables_info, pdf_dimensions):
    # Extract the PDF dimensions
    pdf_width_points, pdf_height_points = pdf_dimensions

    converted_info = {}
    for page_number, page_info in tables_info.items():
        # Parse the image dimensions from the page info
        dimensions = page_info['page_size'].split(' x ')
        ppm_width_pixels = float(dimensions[0])
        ppm_height_pixels = float(dimensions[1])

        scale_factor_width = pdf_width_points / ppm_width_pixels
        scale_factor_height = pdf_height_points / ppm_height_pixels

        converted_tables = []
        for table in page_info['tables']:
            bbox = table['bbox']
            x1_pdf = bbox['x1'] * scale_factor_width
            y1_pdf = pdf_dimensions[1] - bbox['y1'] * scale_factor_height
            x2_pdf = bbox['x2'] * scale_factor_width
            y2_pdf = pdf_dimensions[1] - bbox['y2'] * scale_factor_height

            converted_tables.append({
                'bbox': (x1_pdf, y1_pdf, x2_pdf, y2_pdf)
            })

        converted_info[page_number] = {
            'tables': converted_tables
        }

    return converted_info


pdf_dimensions = (595.276, 793.701)
converted_bboxes = convert_bbox_to_pdf_coords(table_info, pdf_dimensions)
print(converted_bboxes)
```

```python
import camelot

def extract_tables_from_pdf(pdf_path, converted_tables):
    all_extracted_tables = []

    for page_number, page_info in converted_tables.items():

        table_regions = []
        for table in page_info['tables']:
            bbox = table['bbox']
            region = ",".join(map(str, [bbox[0], bbox[1], bbox[2], bbox[3]]))  # x1, y1, x2, y2
            table_regions.append(region)

        if table_regions:
            tables = camelot.read_pdf(pdf_path, flavor="stream", pages=str(page_number), table_regions=table_regions)
            for table in tables:
                all_extracted_tables.append(table.df)

    return all_extracted_tables


extracted_tables = extract_tables_from_pdf(fname, converted_bboxes)
```

```python
import pandas as pd
import numpy as np

import pandas as pd

def adjust_dataframe_headers(dfs):
    """
    Adjusts the headers of each DataFrame in a list. Sets the first row where all elements are not empty and the majority are strings as the new header, and resets the index without adding it to the DataFrame.

    Parameters:
    - dfs (list of pd.DataFrame): List of DataFrames to process.

    Returns:
    - list of pd.DataFrame: List of DataFrames with adjusted headers and without old indexing.
    """
    adjusted_dfs = []  # List to store adjusted DataFrames

    for df in dfs:
        first_valid_index = None
        for idx, row in df.iterrows():
            # Ensure no NaNs and the row is primarily composed of non-empty strings
            if not row.isnull().any() and (row.apply(lambda x: isinstance(x, str) and x != '').sum() >= 2):
                first_valid_index = idx
                break

        if first_valid_index is not None:
            new_header = df.iloc[first_valid_index]  # Assign the new header
            df = df.iloc[first_valid_index + 1:]  # Remove the header row from the data
            df.columns = new_header  # Set the new header
            df.reset_index()  # Reset the index and drop the old one

        adjusted_dfs.append(df)

    return adjusted_dfs



extracted_tables2 = adjust_dataframe_headers(extracted_tables)
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline

model_name_or_path = "TheBloke/Llama-2-13B-chat-GPTQ"
# To use a different branch, change revision
# For example: revision="main"
model = AutoModelForCausalLM.from_pretrained(model_name_or_path,
                                             device_map="auto",
                                             trust_remote_code=False,
                                             revision="main")

tokenizer = AutoTokenizer.from_pretrained(model_name_or_path, use_fast=True)
```

```python
def summarize(table):
  summarizer_instruction = f"""
    <s>[INST]<<SYS>>
      You are a summarizer and your job is to summarize HTM tables and transform them to an understandable text. Your summaries must have the same meaning and not include false information. Make sure you dont use any external knowledge other than whats provided to you. Your final output must only be the summarize in bullet points.
    <</SYS>>
    {table}
    [/INST]
  """
  sequences = pipeline(
      summarizer_instruction,
      do_sample=True,
      top_k=10,
      eos_token_id=tokenizer.eos_token_id,
  )

  answer = sequences[0]['generated_text'].split("[/INST]")[1].strip()
  return answer
```

```python
prompt = extracted_tables2[0]

prompt_template=f'''[INST] <<SYS>>
You are a summarizer and your job is to summarize tables and transform them into text that contains all the information in the table.
DO NOT use any external knowledge, and do not add information that is not mentionned on the table (new words, values ...).
Your summaries must not include false informations.
The next columns contain complementary information to the previous columns, make sure you find links between them.
Your final output must be in bullet points of words that are in the tables (no additions).
<</SYS>>
{prompt}[/INST]

'''

print("\n\n*** Generate:")

input_ids = tokenizer(prompt_template, return_tensors='pt').input_ids.cuda()
output = model.generate(
    input_ids=input_ids,
    temperature=0,      # Lower temperature for less randomness
    do_sample=False,       # Turn off sampling for deterministic output
    top_p=0.99,            # Slightly less restrictive than 0.95           # Allowing a broader selection from the top tokens
    max_new_tokens=1024    # Keeping the output length sufficient
)
print(tokenizer.decode(output[0]))

# Inference can also be done using transformers' pipeline

print("*** Pipeline:")
pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=1024,
    do_sample=False,
    temperature=0,
    top_p=0.95,
)

print(pipe(prompt_template)[0]['generated_text'])
```
