```python
import os
import pandas as pd
from langchain.document_loaders import UnstructuredCSVLoader

def load_all_csv_files(folder_name="tables"):
    docs = []
    delimiters = [';']

    for filename in os.listdir(folder_name):
        print(filename)
        if filename.endswith(".csv"):
            file_path = os.path.join(folder_name, filename)

            for delimiter in delimiters:
                try:

                    loader = UnstructuredCSVLoader(file_path=file_path, mode="elements", delimiter=delimiter)
                    docs.extend(loader.load())
                    print(f'Successfully loaded {filename} with delimiter: {delimiter}')
                    break
                except Exception as e:
                    print(f'Failed to load {filename} with delimiter: {delimiter}. Error: {e}')
            else:
                print(f'Could not determine delimiter for {filename}')
    return docs
```

```python
all_docs = load_all_csv_files("Tables")
print(f"Loaded {len(all_docs)} documents.")

if all_docs:
    print(all_docs[0])
```

```python
from bs4 import BeautifulSoup
from fuzzywuzzy import fuzz

def compare_html_table_with_text(html_table, text, threshold=80):
    soup = BeautifulSoup(html_table, 'html.parser')
    table_text = ' '.join(cell.get_text(strip=True) for cell in soup.find_all(['td', 'th']))

    if not isinstance(table_text, str):
        table_text = str(table_text)
    if not isinstance(text, str):
        text = str(text)

    similarity_score = fuzz.partial_ratio(table_text, text)
    #print(similarity_score)
    return similarity_score >= threshold


similarity = compare_html_table_with_text(all_docs[3].metadata["text_as_html"], texts[14])
print("Similarity:", similarity)
```

```python
from io import StringIO
import pandas as pd

def extract_table_from_text(text):
    lines = text.split("\n")

    table_lines = []
    table_started = False

    for line in lines:
        if "|" in line:
            table_started = True
            table_lines.append(line)
        elif table_started:
            break

    table_text = "\n".join(table_lines)

    table_io = StringIO(table_text)
    df = pd.read_csv(table_io, delimiter="|", skipinitialspace=True)


    df.columns = [col.strip() for col in df.columns]
    df = df[~df.apply(lambda row: row.astype(str).str.contains('----')).any(axis=1)]
    df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)

    df = df.dropna(axis=1, how='all')

    return df
```
